# III. Visualization

### Постройте гистограмму распределения признака 'total_bill'

```python
# ВАШ КОД ЗДЕСЬ
# histplot — это улучшенная гистограмма.
sns.histplot(
    data=tips,      # DataFrame
    x='total_bill', # Столбец для построения
    kde=True        # kde=True - добавляет линию оценки плотности вероятности (Kernel Density Estimate)
)
plt.title('Распределение общего счета');
```

### Постройте scatterplot, представляющий взаимосвязь между признаками 'total_bill' и 'tip'

```python
# ВАШ КОД ЗДЕСЬ
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip'
)
plt.title('Чаевые vs Общий счет');
```

### Постройте pairplot

```python
# ВАШ КОД ЗДЕСЬ
sns.pairplot(
    tips,
    hue='sex' # Раскрашивает точки в зависимости от пола
);
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
# ВАШ КОД ЗДЕСЬ

sns.set_theme(style="whitegrid")
sns.violinplot(
    data=tips,
    x='day',
    y='total_bill',
    hue='time',       # hue - разделяет скрипки по времени
    split=True
)
plt.title('Общий счет по дням недели');
```

### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
# ВАШ КОД ЗДЕСЬ
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(data=tips[tips['time'] == 'Lunch'], x='tip', kde=True,ax=axes[0],color='purple')
sns.histplot(data=tips[tips['time'] == 'Dinner'], x='tip', kde=True,ax=axes[1])
axes[0].set_title('Распределение чаевых - Lunch')
axes[1].set_title('Распределение чаевых - Dinner')
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
# ВАШ КОД ЗДЕСЬ
# для наглядности - на одном графике но разным цветом
sns.scatterplot(
    data=tips,
    x='total_bill',
    y='tip',
    hue='sex',   # hue - цвет точек зависит от переменной 'sex'
    style='smoker'    # style - форма маркера зависит от категориальной 'smoker'
)
plt.title('Чаевые vs Общий счет');
#sns.regplot(data=tips, x='total_bill', y='tip')
```

## Сделайте выводы по анализу датасета и построенным графикам. По желанию можете продолжить анализ данных и также отразить это в выводах.

```python
# ВАШ КОД ЗДЕСЬ
tips = sns.load_dataset('tips')
tips_corr = tips.copy()
day_mapping = {'Thur': 1, 'Fri': 2, 'Sat': 3, 'Sun': 4}
tips_corr['day_num'] = tips_corr['day'].map(day_mapping)

print(tips_corr.head())
numeric_cols = tips_corr[['total_bill', 'day_num']]
#print(numeric_cols)
correlation_matrix = numeric_cols.corr()
print("Матрица корреляции (total_bill vs day_num):")
print(correlation_matrix)
print(f"\nКорреляция между total_bill и днем недели: {correlation_matrix.loc['total_bill', 'day_num']:.3f}")

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0,
            fmt='.3f', square=True, linewidths=1, cbar_kws={"shrink": 0.8})
plt.title('Корреляция между total_bill и днем недели', fontsize=14, fontweight='bold')
plt.show()
#Как показано, корреляция между днем недели и общим счетом слабоинформативна
#поэтому,для визуализации связи между категориальным и числовым признаками использовали не матрицу корреляции,а violin plot
##Общий вывод: мужчины оставляют больше чаевых, чем женщины, ъотя урение делает послежних добрее;)
```
