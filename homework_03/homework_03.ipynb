## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот.

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
# ВАШ КОД ЗДЕСЬ
def convert(text, frmt=None):
    if not isinstance(text, str) or not text:
        return text

    # Определяем исходный формат
    if frmt == 'snake':
        target = 'snake'
    elif frmt == 'pascal':
        target = 'pascal'
    else:
        if ('_' in text and text.islower()):
            target = 'pascal'
        elif (text[0].isupper() and '_' not in text):
            target = 'snake'
        else:
            target = 'pascal' if '_' in text else 'snake'

    # Конвертация в PascalCase
    if target == 'pascal':
        words = text.split('_')
        result = ''
        for word in words:
            if word:
                result += word[0].upper() + word[1:].lower()
        return result

    # Конвертация в snake_case
    else:  # target == 'snake'
        result = ''
        i = 0
        for char in text:
          if i > 0 and char.isupper():
            result += '_' + char.lower()
          else:
            result += char.lower()
          i += 1
        return result
#print(convert("otus_course"))
#print(convert("PythonIsTheBest"))
```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
# ВАШ КОД ЗДЕСЬ
def is_(n):

  #проверка на целое число
  if not isinstance(n,(int,float)):
    return False
  if isinstance(n,float) and n%1!=0:
    return False

  return (n>1) and all(n%i!=0 for i in range(2,int(n**0.5)+1))

#print(is_(25))
```
